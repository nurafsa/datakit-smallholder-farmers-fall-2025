```
/* Create new modified Producers Direct table with expanded QuestionTopic and ResponseTopic fields */
CREATE MULTISET TABLE ADLMKT_Teradata_Cares.ProducersDirectMod_OFS ,NO FALLBACK ,
     MAP = TD_MAP1,
     STORAGE = TD_OFSSTORAGE 
     (
      QuestionID INTEGER,
      QuestionUserID INTEGER,
      QuestionLanguage VARCHAR(100) CHARACTER SET UNICODE NOT CASESPECIFIC,
      QuestionContent VARCHAR(5000) CHARACTER SET UNICODE NOT CASESPECIFIC,
      QuestionTopic VARCHAR(1000) CHARACTER SET UNICODE NOT CASESPECIFIC,
      QuestionSentDT TIMESTAMP(6) WITH TIME ZONE,
      ResponseID INTEGER,
      ResponseUserID INTEGER,
      ResponseLanguage VARCHAR(100) CHARACTER SET UNICODE NOT CASESPECIFIC,
      ResponseContent VARCHAR(5000) CHARACTER SET UNICODE NOT CASESPECIFIC,
      ResponseTopic VARCHAR(1000) CHARACTER SET UNICODE NOT CASESPECIFIC,
      ResponseSentDT TIMESTAMP(6) WITH TIME ZONE,
      QuestionUserType VARCHAR(100) CHARACTER SET UNICODE NOT CASESPECIFIC,
      QuestionUserStatus VARCHAR(25) CHARACTER SET UNICODE NOT CASESPECIFIC,
      QuestionUserCountryCode CHAR(2) CHARACTER SET UNICODE NOT CASESPECIFIC,
      QuestionUserGender VARCHAR(25) CHARACTER SET UNICODE NOT CASESPECIFIC,
      QuestionUserBirthDate DATE FORMAT 'YYYY/MM/DD',
      QuestionUserCreatedDT TIMESTAMP(6) WITH TIME ZONE,
      ResponseUserType VARCHAR(100) CHARACTER SET UNICODE NOT CASESPECIFIC,
      ResponseUserStatus VARCHAR(25) CHARACTER SET UNICODE NOT CASESPECIFIC,
      ResponseUserCountryCode CHAR(2) CHARACTER SET UNICODE NOT CASESPECIFIC,
      ResponseUserGender VARCHAR(25) CHARACTER SET UNICODE NOT CASESPECIFIC,
      ResponseUserBirthDate DATE FORMAT 'YYYY/MM/DD',
      ResponseUserCreatedDT TIMESTAMP(6) WITH TIME ZONE)
ORDER BY ( QuestionLanguage )
PRIMARY INDEX ( QuestionID )
PARTITION BY COLUMN;
```
